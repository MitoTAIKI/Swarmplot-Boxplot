```python
#@markdown ###Step 2: エクセルファイルを読み込む
Path = "/content/Data_sample.xlsx" #@param {type:"string"}
input_book = pd.ExcelFile(Path)
```

```python
#@markdown ###Step 3: エクセルファイルのシート数とシート名を表示
input_sheet_name = input_book.sheet_names
num_sheet_name = input_book.sheet_names
num_sheet = len(input_sheet_name)
print("Sheetの数 =", num_sheet)
print(input_sheet_name)
```

```python
#@markdown ###Step 4 シートを選択 

sheet_number = 0 #@param {type:"string"}
input_sheet_df = input_book.parse(input_sheet_name[sheet_number])
print("Sheet name:", input_sheet_name[sheet_number])
input_sheet_df.head(10000)
```

```python
#@markdown ###Step 5 グラフ作成

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
Dataset = input_sheet_df.head(10000)
data_name = "Parameter"
group_name = "Group"

#@markdown グラフのサイズと解像度 (デフォルト設定: サイズ = 4:5, 解像度 dpi = 300)
x = 4 #@param {type:"number"}
y = 5 #@param {type:"number"}
d = 300 #@param {type:"number"}
fig = plt.figure(figsize=(x,y), dpi=d)
ax = fig.add_subplot() 

#Seaborn swarmplot (Dot plot)
sns.swarmplot(x=Dataset[group_name], y=Dataset[data_name], data=Dataset, palette=sns.color_palette(["silver","silver","silver"],3), alpha=0.7, linewidth=0.5, edgecolor="gray")
#Seaborn boxplot
bp = sns.boxplot(x=Dataset[group_name], y=Dataset[data_name], data=Dataset, palette=sns.color_palette(["yellowgreen","yellowgreen","yellowgreen"],3), linewidth=1.0, sym="")
plt.title(group_name+" - "+data_name, size=10)
plt.ylabel(data_name)
ax.yaxis.grid(alpha=0.4)

#@markdown 縦軸の範囲
y_min = 0 #@param {type:"number"}
y_max = 80 #@param {type:"number"}
ax.set_ylim(y_min, y_max)

#@markdown 縦軸の目盛り間隔
y_step = 20 #@param {type:"number"}
plt.yticks(np.arange(y_min, y_max+1, step=y_step))

#@markdown 縦軸のラベル
y_label = "" #@param {type:"string"}
font_size = 10 #@param {type:"number"}
ax.set_ylabel(y_label, fontsize=font_size)

#横軸のラベル
ax.set_xlabel("")

#@markdown グラフのタイトル
graph_title = "" #@param {type:"string"}
ax.set_title(graph_title)

#作成したグラフを表示
plt.show()
```
